# Lists and Mutability

```python
x = 34
y = x
y = 20

print(f"x is = {x} and y is = {y}")
```

```python
mammals = ['cat', 'kangaroo', 'hourse']
fish    = ['tuna', 'shark', 'catfish']
```

```python
animals = [mammals, fish]
```

```python
print(animals)
```

```python
mammals[0] = 'mule'
print(animals)
```

```python
mammals_2 = mammals
mammals_2[0] = 'dog'

print(mammals_2)
print(mammals)
print(animals)
```

#### So it is the same list with different names but can't we make the copy of the list?

```python
mammals_3 = mammals[:]
mammals_3[0] = 'mouse'

print(mammals_3)
print(mammals)
print(animals)
```

```python
s1 = ['student 1', 'James', 'Brown']
s2 = ['student 2', 'Chris', 'Black']
s3 = ['student 3', 'Phil', 'Blue']
```

```python
students = [s1, s2, s3]
print(students)
```

```python
students_copy = students[:] # called shallow copy
```

```python
print(students_copy)
```

```python
students_copy[0][0] = 'student 1 changed'
print(students_copy)
print(students)
```

# <b> Deep Copying </b>

```python
import copy
```

```python
student_3 = copy.deepcopy(students)
```

```python
student_3[0][0] = 'student 1 changed again'
print(student_3)
print('---------')
print(students)
```

# <b> Passing lists as argument </b>

```python
def change_var(x):
    x = x + 1

def change_lst(lst):
    lst.append(24)
```

```python
x = 25
change_var(x)
print(x)
```

```python
lst = [2, 3, 4, 5]
change_lst(lst)
print(lst)
```
